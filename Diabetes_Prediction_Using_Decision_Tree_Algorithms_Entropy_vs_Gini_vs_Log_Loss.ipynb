# ***Decision Tree workout***

```python
import numpy as np
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/Dataset ML/diabetes.csv")
data
```

```python
data.isna().sum()
```

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion="entropy")
model.fit(x_train,y_train)
y_pred= model.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
model1=DecisionTreeClassifier(criterion="gini")
model1.fit(x_train,y_train)
y_pred= model1.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
model2=DecisionTreeClassifier(criterion="log_loss")
model2.fit(x_train,y_train)
y_pred= model2.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
print (classification_report(y_test,y_pred))
```
